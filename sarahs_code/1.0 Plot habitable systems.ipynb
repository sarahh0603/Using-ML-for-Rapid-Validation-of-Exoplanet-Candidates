```python
import matplotlib.pyplot as plt
import numpy as np

# Data for each system
systems = [
    {
        "name": "Solar System",
        "radii": {
            "Mercury": 0.387,
            "Venus": 0.723,
            "Earth": 1.0,
            "Mars": 1.524
        },
        "habitable_zone": {
            "Inner": 0.953,
            "Outer": 1.374
        },
        "star_color": "yellow",
        "star_label": "Sun",
        "title": "a)"
    },
    {
        "name": "TIC149302744",
        "radii": {
            "699.01": 0.127,
            "699.02": 0.219,
            "699.03": 1.614
        },
        "habitable_zone": {
            "Inner": 1.507,
            "Outer": 2.548
        },
        "star_color": "#D8EFFF",
        "star_label": "TIC149302744",
        "title": "b)"
    },
    {
        "name": "TIC306472057",
        "radii": {
            "791.01": 0.587,
            "791.02": 2.077
        },
        "habitable_zone": {
            "Inner": 1.856,
            "Outer": 3.085
        },
        "star_color": "#D8EFFF",
        "star_label": "TIC306472057",
        "title": "c)"
    }
]

# Create figure with 3 subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 6), facecolor='white')

# Determine maximum radius across all systems for consistent scaling
max_orbit = max(max(sys["radii"].values()) for sys in systems)
max_hz = max(sys["habitable_zone"]["Outer"] for sys in systems)
max_radius = max(max_orbit, max_hz) * 1.2

def create_system_plot(ax, system):
    radii = system["radii"]
    hz = system["habitable_zone"]
    
    # Set equal aspect ratio and fixed scale
    ax.set_aspect('equal')
    ax.set_xlim(-max_radius, max_radius)
    ax.set_ylim(-max_radius, max_radius)
    
    # Shade habitable zone in green
    ax.add_patch(plt.Circle((0, 0), hz["Outer"], color='green', alpha=0.2, fill=True))
    ax.add_patch(plt.Circle((0, 0), hz["Inner"], color='white', alpha=1.0, fill=True))
    
    # Plot star
    ax.scatter(0, 0, color=system["star_color"], s=300, label=system["star_label"], edgecolor='black')
    
    # Plot orbits
    for planet, radius in radii.items():
        orbit = plt.Circle((0, 0), radius, color='blue', fill=False, linestyle='--', linewidth=1.5)
        ax.add_artist(orbit)
        
        # Position labels with alternating placement
        angle = 45 if list(radii.keys()).index(planet) % 2 else -45
        x = radius * np.cos(np.radians(angle))
        y = radius * np.sin(np.radians(angle))
        offset = 15 if y > 0 else -15
        ax.annotate(planet, xy=(x, y), xytext=(0, offset),
                    textcoords='offset points', ha='center', 
                    va='bottom' if y > 0 else 'top', fontsize=18, 
                    color='blue')
      
    # Add legend
    ax.legend(loc='upper right', fontsize=25, frameon=False)

# Create each plot
for i, system in enumerate(systems):
    create_system_plot(axs[i], system)

# Adjust layout with minimal space between subplots
plt.subplots_adjust(wspace=0.05)

# Save and show
plt.savefig("uniform_scale_systems_with_hz.png", bbox_inches='tight', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the radii of the orbits relative to Earth's orbit (1 AU) for the first plot
radii_1 = {
    "Mercury": 0.387,
    "Venus": 0.723,
    "Earth": 1.0,
    "Mars": 1.524
}

# Define the Habitable Zone radii for the first plot
habitable_zone_1 = {
    "Inner": 0.953,
    "Outer": 1.374
}

# Define the radii of the orbits relative to Earth's orbit (1 AU) for the second plot
radii_2 = {
    "699.01": 0.127,
    "699.02": 0.219,
    "699.03": 1.614
}

# Define the Habitable Zone radii for the second plot
habitable_zone_2 = {
    "Inner": 1.507,
    "Outer": 2.548
}

# Define the radii of the orbits relative to Earth's orbit (1 AU) for the third plot
radii_3 = {
    "791.01": 0.587,
    "791.02": 2.077
}

# Define the Habitable Zone radii for the third plot
habitable_zone_3 = {
    "Inner": 1.856,
    "Outer": 3.085
}

# Create a figure and axes for three subplots in one row
fig, axs = plt.subplots(1, 3, figsize=(18, 6))  # 1 row, 3 columns

# Function to create a plot
def create_plot(ax, radii, habitable_zone, title, star_color, star_label):
    # Set the aspect ratio to be equal
    ax.set_aspect('equal')

    # Shade the area between the inner and outer circles
    ax.add_patch(plt.Circle((0, 0), habitable_zone["Outer"], color='green', alpha=0.2, fill=True))
    ax.add_patch(plt.Circle((0, 0), habitable_zone["Inner"], color='white', alpha=1.0, fill=True))

    # Plot the Sun at the center
    ax.scatter(0, 0, color=star_color, s=200, label=star_label)

    # Plot the orbits of the planets
    for planet, radius in radii.items():
        circle = plt.Circle((0, 0), radius, color='blue', fill=False, linestyle='--')
        ax.add_artist(circle)

    # Set the limits of the plot
    max_radius = max(max(radii.values()), habitable_zone["Outer"])
    ax.set_xlim(-max_radius * 1.1, max_radius * 1.1)
    ax.set_ylim(-max_radius * 1.1, max_radius * 1.1)

    # Remove X and Y axes
    ax.set_xticks([])  # Remove X-axis tick marks
    ax.set_yticks([])  # Remove Y-axis tick marks
    for spine in ax.spines.values():  # Remove axis spines (lines)
        spine.set_visible(False)

    # Add title at the upper left corner (within plot limits)
    ax.text(-max_radius * 1.05, max_radius * 1.05, title, fontsize=25, ha='left', va='top', color='black')

    # Add a legend at the upper right corner, aligned with the title
    handles, labels = ax.get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    ax.legend(by_label.values(), by_label.keys(), loc='upper right', bbox_to_anchor=(1.0, 1.0), fontsize=25, frameon=False)

    # Add call-out labels for each planet with alternating positions
    positions = ['top', 'bottom', 'top', 'bottom']
    for i, planet in enumerate(radii):
        if positions[i % len(positions)] == 'top':
            ax.annotate(planet, xy=(radii[planet], 0), xytext=(0, 10), textcoords='offset points', ha='center', va='bottom', fontsize=15, color='blue')
        else:
            ax.annotate(planet, xy=(radii[planet], 0), xytext=(0, -10), textcoords='offset points', ha='center', va='top', fontsize=15, color='blue')

# Create each plot using the function
create_plot(axs[0], radii_1, habitable_zone_1, "a)", 'yellow', 'Sun')
create_plot(axs[1], radii_2, habitable_zone_2, "b)", '#D8EFFF', 'TIC149302744')
create_plot(axs[2], radii_3, habitable_zone_3, "c)", '#D8EFFF', 'TIC306472057')

# Adjust layout to keep same scale for all plots
for ax in axs:
    ax.set_xlim(-3 * 1.1, 3 * 1.1)
    ax.set_ylim(-3 * 1.1, 3 * 1.1)

# Save the figure to a PDF file
plt.savefig("ToScale_habitable_systems.pdf", format="pdf", bbox_inches="tight")

# Show all plots in one row
plt.tight_layout()
plt.show()
```
